```python
import pandas as pd
import folium
```

```python
# Створення DataFrame з даними
data = {
    'State': ['China', 'France', 'India', 'Israel', 'Japan', 'Korea, North', 'Pakistan', 'Russia', 'UK', 'USA'],
    'Facility': ['Jiuquan pilot plant', 'La Hague UP2', 'Kalpakkam', 'Dimona', 'JNC Tokai', 'Yongbyon', 'Chashma', 'Mayak RT-1', 'Sellafield B205', 'H-canyon, Savannah River Site'],
    'Latitude': [40.0, 49.4, 12.6, 31.0, 35.0, 40.6, 33.6, 54.9, 54.4, 33.2],
    'Longitude': [100.5, -0.5, 80.0, 34.8, 138.0, 124.8, 72.8, 60.0, -3.0, -81.2]
}

df = pd.DataFrame(data)
```

```python
# Створення інтерактивної карти
m = folium.Map(location=[20, 0], zoom_start=2)

# Додавання маркерів для кожного підприємства
for i in range(len(df)):
    folium.Marker(
        location=[df['Latitude'][i], df['Longitude'][i]],
        popup=f"{df['Facility'][i]} in {df['State'][i]}",
        icon=folium.Icon(color='blue')
    ).add_to(m)

# Відображення карти
m
```
