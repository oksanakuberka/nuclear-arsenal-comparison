# Importing Libraries and Creating a Bar Plot of Nuclear Arsenals


The following code imports the `matplotlib.pyplot` and `seaborn` libraries, sets the style for the plot, and creates a bar plot to visualize the total inventory of nuclear warheads by country.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style of the plot
sns.set(style="whitegrid")

# Creating a bar plot for the total inventory of nuclear warheads by country
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total inventory', data=df, palette='viridis')
plt.title('Comparison of Nuclear Arsenals by Country')
plt.xlabel('Country')
plt.ylabel('Total Nuclear Warheads')
plt.xticks(rotation=45)
plt.show()


```

# Finding the Country with the Largest Nuclear Arsenal

The following code finds the country with the largest nuclear arsenal by using the `idxmax()` method to get the index of the maximum value in the `Total inventory` column. It then prints the country's name and the total number of nuclear warheads it possesses.

```python
max_country = df.loc[df['Total inventory'].idxmax()] # Find the country with the largest nuclear arsenal using the idxmax() method
print(f"The country with the largest nuclear arsenal is {max_country['Country']} with {max_country['Total inventory']} warheads.") # Print information about the country with the largest arsenal
```
